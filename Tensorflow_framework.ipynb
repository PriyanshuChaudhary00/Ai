```python
import numpy as np
import tensorflow as tf
```

```python
w = tf.Variable(0 , dtype = tf.float32)
optimizer = tf.keras.optimizers.Adam(0.1)

def train_step():
  with tf.GradientTape() as Tape:
    cost = w ** 2 + 10 * w + 100
  trainable_variable  = [w]
  grads = Tape.gradient(cost , trainable_variable)
  optimizer.apply_gradients(zip(grads , trainable_variable))


print(w)

for i in range(1000):
  train_step()
print(w)
```

```python
w = tf.Variable(0 , dtype = tf.float32)
x = np.array([1.0 , -10.0 , 25.0] , dtype = np.float32)
optimizer = tf.keras.optimizers.Adam(0.1)
def Training(x , w , optimizer):
  def cost_fn():
    return x[0] * w ** 2 + x[1] * 10 * w + x[2] * 100

  for i in range(1000):
    with tf.GradientTape() as tape:
      current_cost = cost_fn()

    gradients = tape.gradient(current_cost, [w])

    optimizer.apply_gradients(zip(gradients , [w]))
  return w

print("Initial w:", w.numpy())

print("Final w:", Training(x , w , optimizer=optimizer))
```
