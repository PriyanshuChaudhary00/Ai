```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
from torch.nn.modules import linear
# Distance
distances = torch.tensor([[1],[2],[3],[4]] , dtype = torch.float32)
# Time
times = torch.tensor([[6.96] , [12.36] , [16.77] , [22.21]])
```

```python
# Model
model = nn.Sequential(nn.Linear(1 , 1))
```

```python
# define loss fn amd optimizer

loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters() , lr=0.01)
```

```python
for epoch in range(500):
  # reset optemizer to 0
  optimizer.zero_grad()

  # make prediction
  output = model(distances)

  # calculate loss
  loss = loss_fn(output , times)

  # calculate adjustment
  loss.backward()

  # update model
  optimizer.step()
  if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch + 1}: Loss = {loss.item()}")
```

```python
with torch.no_grad():
  test_distance = torch.tensor([[5.0]] , dtype = torch.float32)
  prediction = model(test_distance)
  print("prediction is ", prediction)
```

```python
# Access the first (and only) layer in the sequential model
layer = model[0]

# Get weights and bias
weights = layer.weight.data.numpy()
bias = layer.bias.data.numpy()

print(f"Weight: {weights}")
print(f"Bias: {bias}")
```
